```python
!pip install japanize-matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib
from sklearn import model_selection, preprocessing, linear_model, svm
import sklearn

from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("train.csv", index_col=0) #学習用データ
test = pd.read_csv("test.csv", index_col=0) #評価用データ
sample_submit = pd.read_csv("sample_submission.csv", index_col=0, header=None) #応募用サンプルファイル
```

```python
train.head() #headで構造確認
```

```python
#yに目的変数、xにそれ以外を
x = train.iloc[:, 0:-1]
y = train.iloc[:, -1]
```

```python
#目的変数が消えていることの確認
x.head()
```

```python
#xとyをそれぞれ学習用とテスト用に分割
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, train_size=0.8)
```

```python
#学習モデルの選択
#反復回数80000くらいが良さそう
#80000の時 0.3958333333333333
model = LinearSVC(max_iter=80000)
#モデルの学習
model.fit(x_train, y_train)
#予測モデルの作成
pred = model.predict(x_test)
```

```python
print(accuracy_score(y_test, pred))
```
